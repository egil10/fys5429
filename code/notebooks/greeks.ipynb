# Greeks

```python
# Function to compute greeks
def compute_greeks(model, S, tau, r=None, sigma=None, q=None):
    """
    Compute all Greeks from a trained PINN via autograd.

    S and tau must be tensors with requires_grad=True.
    r, sigma, q are optional — pass as requires_grad=True tensors
    to enable their Greeks, otherwise those return None.
    """

    V = model(S, tau)

    # First order
    Delta = torch.autograd.grad(V, S, create_graph=True)[0]
    Theta = torch.autograd.grad(V, tau, create_graph=True)[0]

    Rho = torch.autograd.grad(V, r, create_graph=True)[0] if r is not None and r.requires_grad else None
    Vega = torch.autograd.grad(V, sigma, create_graph=True)[0] if sigma is not None and sigma.requires_grad else None
    Phi = torch.autograd.grad(V, q, create_graph=True)[0] if q is not None and q.requires_grad else None

    # Second order
    Gamma = torch.autograd.grad(Delta, S, create_graph=True)[0]
    Charm = torch.autograd.grad(Delta, tau, create_graph=True)[0]

    Vomma = torch.autograd.grad(Vega, sigma, create_graph=True)[0] if Vega is not None else None
    Vanna = torch.autograd.grad(Delta, sigma, create_graph=True)[0] if sigma is not None and sigma.requires_grad else None

    # Third order
    Speed = torch.autograd.grad(Gamma, S, create_graph=True)[0]
    Color = torch.autograd.grad(Gamma, tau, create_graph=True)[0]
    Zomma = torch.autograd.grad(Gamma, sigma, create_graph=True)[0] if sigma is not None and sigma.requires_grad else None

    return {
        "V": V,
        "Delta": Delta,
        "Theta": Theta,
        "Rho": Rho,
        "Phi": Phi,
        "Vega": Vega,
        "Gamma": Gamma,
        "Vomma": Vomma,
        "Vanna": Vanna,
        "Charm": Charm,
        "Speed": Speed,
        "Zomma": Zomma,
        "Color": Color,
    }
```
